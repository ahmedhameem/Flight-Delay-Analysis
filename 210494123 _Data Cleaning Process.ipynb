# Data Cleaning Process



```python
#import the required libraries
import pandas as pd
import numpy as  np
```

```python
# import the dataframes used for analysis
df_06=pd.read_csv('2006.csv')
df_07=pd.read_csv('2007.csv')

#join both dataframes into one dataframe
df=pd.concat([df_06,df_07])
```

```python
df # explore the combined dataset
```

```python
#describe the dataframe
df.describe()
```

```python
 # check for null values
pd.isnull(df).sum(axis=0)
```

```python
df=df.drop(columns="CancellationCode")  #98% of the entries in "CancellationCode" is missing therefore it is better to drop the
                                        # column than removing the rows with null values as this will result in a massive loss of data
```

```python
# checking for null values after removing column 'CancellationCode'
pd.isnull(df).sum(axis=0)
```

```python
# now we remove the rows of the dataset with null values
# if we removed the rows along with the column 'CancellationCode' we would not be left with a lot of data
df=df.dropna()
df
```

```python
# rechecking for null values in the dataframe
pd.isnull(df).sum(axis=0)
```

```python
df.describe()
```

```python
for i in df[1:]:
    print(f'Colname: {i},min :{df[i].min()}, max :{df[i].max()}')
# we are trying to find any utliers that may exist in the dataset 
```

```python
# we can see columns such as DepTime & ArrTime have max values that exceed the typical 24 hour day time period and this does not make sense. 
# the values that exceed the 24 hour standard time period must be removed as it is misleading to have such data .
# the max time for these columns can be 2359 and minimum 1
df = df[(df["DepTime"]<2400) & (df["ArrTime"] <2400)]

```

```python
## check the max of Deptime & ArrTime
# now   all times are less than or equal to 2359
print(df["DepTime"].max())
print(df["ArrTime"].max())
```

```python
for i in df[1:]:
    print(f'Colname: {i},min :{df[i].min()}, max :{df[i].max()}')
    
    # the max and min values for cancelled and diverted flights is 0 indicating
    #all data with regard to cancelled and diverted flights have been removed
```

```python
# saving the cleaned dataset to be used as a csv for tasks to be carried out
df.to_csv('clean_data.csv')
```
